### Hierarchical visualisation

```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from scipy.spatial.distance import squareform
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (12, 7))
dissimilarity = 1 - abs(prices.corr())
Z = linkage(squareform(dissimilarity), 'ward')

dendrogram(Z, labels=prices.columns, leaf_rotation=90, color_threshold=0.8 )
plt.savefig('tree_.png', dpi=500);
```

```python
threshold = 1.5
labels = fcluster(Z, threshold, criterion='distance')

# Show the cluster
labels
```

### Network Visualisation

```python
import networkx as nx
```

```python
links = prices.corr().stack().reset_index()
links.columns = ['var1', 'var2', 'value']
```

```python
links_filtered=links.loc[ (links['value'] > 0.7) & (links['var1'] != links['var2']) ]
G=nx.from_pandas_edgelist(links_filtered, 'var1', 'var2', edge_attr = 'value')
nx.draw(G, with_labels=True, node_color='orange', node_size=300, edge_color='black', linewidths=1, font_size=10)
```

```python
from pyvis.network import Network
```

```python
net = Network(width='900px', height='600px', notebook=True)
net.barnes_hut(spring_strength=0.02, 
                                  damping=0.18, 
                                  gravity=-800, 
                                  spring_length=200)
net.from_nx(G)
for e in net.edges:
    e['weight'] = (5*e['value'])**5
    
net.show('test.html')
```

## Daily price changes

```python
prices = pd.read_csv('top_crypto_3y.csv')
prices['Date'] = pd.to_datetime(prices['Date'])
prices = prices.set_index('Date')
```

```python
prices
```

```python
daily_pct = pd.DataFrame()
for col in prices.columns:
    daily_pct[col] = ((prices[col] - prices[col].shift(1))/prices[col])*100
```

```python
daily_pct = daily_pct.dropna()
```

```python
daily_pct.corr()['BTC-USD'].sort_values()
```

```python
plt.figure(figsize = (12, 7))
dissimilarity = 1 - daily_pct.corr()
Z = linkage(squareform(dissimilarity), 'ward')

dendrogram(Z, labels=prices.columns, leaf_rotation=90, color_threshold=0.6 );
```

```python
daily_pct.columns = [x[:-4] for x in daily_pct.columns]
```

```python

links = daily_pct[['BTC', 'ETH', 'USDT', 'BNB', 'SOL', 'XRP', 'USDC', 'DOGE',
       'ADA', 'AVAX', 'SHIB', 'DOT', 'BCH', 'LINK', 'TRX', 'MATIC',
       'ICP', 'UNI7083', 'NEAR', 'LTC', 'LEO', 'DAI', 'STX4847', 'FIL',
       'ATOM']].corr().stack().reset_index()
links.columns = ['var1', 'var2', 'value']
links_filtered=links.loc[ (links['value'] > 0.8) & (links['var1'] != links['var2']) ]
G=nx.from_pandas_edgelist(links_filtered, 'var1', 'var2', edge_attr = 'value')

net = Network(width='900px', height='600px', notebook=True,bgcolor='#333333', font_color='#000000')
net.barnes_hut(spring_strength=0.03, 
                                  damping=0.18, 
                                  gravity=-800, 
                                  spring_length=100)
net.from_nx(G)
for e in net.edges:
    e['weight'] = e['value']**3
    e['label'] = e['value']
    
net.show('test.html')
```
